**Multiple Linear Regression:**

**Definition:**
Multiple Linear Regression is a statistical method that models the relationship between multiple independent variables (features) and a single dependent variable (target) by fitting a linear equation to observed data. It extends the concept of simple linear regression to more than one predictor variable.

**Key Points:**

1. **Multiple Predictors:** In multiple linear regression, there are more than one independent variable influencing the dependent variable.

2. **Linear Relationship:** The relationship between the independent variables and the dependent variable is assumed to be linear. This means that a change in any of the independent variables is associated with a constant change in the dependent variable.

3. **Regression Equation:** The multiple linear regression equation can be written as follows:
\[ y = b_0 + b_1x_1 + b_2x_2 + \ldots + b_nx_n + \varepsilon \]
where \( y \) is the dependent variable, \( x_1, x_2, \ldots, x_n \) are the independent variables, \( b_0 \) is the intercept, \( b_1, b_2, \ldots, b_n \) are the coefficients, and \( \varepsilon \) represents the error term.

4. **Interpretation of Coefficients:** The coefficients (\( b_0, b_1, b_2, \ldots, b_n \)) represent the change in the dependent variable for a one-unit change in the corresponding independent variable, holding other variables constant.

5. **Ordinary Least Squares (OLS):** The most common method for estimating the coefficients in multiple linear regression is OLS. It minimizes the sum of squared differences between observed and predicted values.

6. **Assumptions:** Multiple linear regression assumes linearity, independence of errors, homoscedasticity (constant variance of errors), and absence of multicollinearity (correlation between independent variables).

7. **Evaluation Metrics:** Common evaluation metrics for multiple linear regression include Mean Squared Error (MSE), Root Mean Squared Error (RMSE), and R-squared (\( R^2 \)) which measures the proportion of the variance in the dependent variable that is predictable from the independent variables.

8. **Feature Selection:** Choosing relevant features and avoiding multicollinearity is crucial. Feature selection techniques like backward elimination, forward selection, or regularization methods (e.g., Lasso and Ridge regression) can be applied.

9. **Use Cases:** Multiple linear regression is used in various fields such as economics, finance, biology, and social sciences to analyze and predict relationships between variables.

10. **Caution:** Care must be taken to ensure the assumptions of multiple linear regression are met. Additionally, the model's predictions should be interpreted cautiously and within the context of the problem domain.

```python
import pandas as pd

dataset = pd.read_csv("admissiondata.csv") # Read the CSV file into a DataFrame

x = dataset.iloc[:, 0:6].values # Selecting features (first 6 columns) for x and the 7th column as the target for y
y = dataset.iloc[:,0:7].values  # Corrected index to select the 7th column as the target
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
# x: features, y: target variable
# test_size=0.2: 20% of the data will be used for testing, and 80% for training

x_train, x_test, y_train , y_test = train_test_split(x,y,test_size = 0.2)

# Now, x_train and y_train contain the features and target values for training respectively.
# Similarly, x_test and y_test contain the features and target values for testing respectively.
```

```python
from sklearn.preprocessing import PolynomialFeatures

multi_poly = PolynomialFeatures(degree = 2)
x_poly = multi_poly.fit_transform(x_train)
multi_poly.fit(x_poly,y_train)
```

```python
from sklearn.linear_model import LinearRegression

lin_reg_multi = LinearRegression()
lin_reg_multi.fit(x_poly, y_train)
```

```python
y_preds = lin_reg_multi.predict(multi_poly.fit_transform(x_test))

from sklearn import metrics
print(metrics.mean_squared_error(y_test,y_preds))
```
