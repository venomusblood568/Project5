```python
import keras #ml library
import numpy as np #mathematics library
import pandas as pd #data_handling library
import matplotlib.pyplot as plt #used as graph
```

```python
data = pd.read_csv("IowaHousingPrices.csv")
```

```python
square_feet = data[["SquareFeet"]].values
sales_price = data[["SalePrice"]].values
```

```python
model = keras.Sequential()
model.add(keras.layers.Dense(1, input_shape=(1,)))
model.compile(keras.optimizers.Adam(lr=1), 'mean_squared_error')

model.fit(square_feet,sales_price, epochs=50, batch_size=2)
```

```python
#Plot datapoints
data.plot(kind='scatter',
       x='SquareFeet',
       y='SalePrice', title='Housing Prices and Square Footage of Iowa Homes')


y_pred = model.predict(square_feet) #The predicted housing price based on square feet

#Plot the linear regression line
plt.plot(square_feet, y_pred, color='red')
```

```python
import numpy as np

# Assuming newSF is a single square feet value you want to predict for
newSF = 2000

# Reshape the input data to have shape (1, 1)
input_data = np.array([[newSF]])

# Make a prediction
prediction = model.predict(input_data)

# Print the prediction
print("Predicted sale price for", newSF, "square feet:", prediction[0][0])
```
