cancer cells can be :

malignant : tumor is cancerious so in the dataset we will assign this a 1

Or

bening : tumor is not cancerious and this as 0

```python
import pandas as pd

dataset = pd.read_csv("cancer.csv")
```

```python
 x = dataset.drop(columns=["diagnosis(1=m, 0=b)"])
```

```python
y = dataset["diagnosis(1=m, 0=b)"]
```

```python
from sklearn.model_selection import train_test_split

x_train , _test , y_train , y_test = train_test_split(x,y,test_size=0.2)
```

```python
import tensorflow as tf

model = tf.keras.models.Sequential()
```

```python
model.add(tf.keras.layers.Dense(256, input_shape=x_train.shape[1:], activation='sigmoid'))
model.add(tf.keras.layers.Dense(256, activation='sigmoid'))
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer="adam",loss = "binary_crossentropy",metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs = 1500)
```

```python
model.evaluate(x_train,y_train)
```
